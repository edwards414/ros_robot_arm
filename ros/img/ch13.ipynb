## 13.4微積分

```python
#11
#a
x,y = symbols('x,y')
ex = sqrt(x**2+3*x+2)
#a
lim = Limit(ex,x,0,dir='+-');lim.doit()
#b
ex = (x**2-7*x+10)/(x+2)
lim = Limit(ex,x,2,dir='-');lim.doit()
#c
ex = sin(x)/(2*x-1)
lim = Limit(ex,x,oo,dir='-');lim.doit()
#d
ex = ((x+3)*(x-4)*(2*x+6))/(x**3-3)
lim = Limit(ex,x,oo,dir='-');lim.doit()
#e
ex = x**x
lim = Limit(ex,x,0,dir='-');lim.doit()
#f
ex = 1/(x-2)
lim = Limit(ex,x,2,dir='-');lim.doit()
```

```python
#12
#a
ex = 1/(x**(3/2)+1)
Derivative(ex,x).doit()
#b
ex = (4*x-3)/sin(x+1)
Derivative(ex,x).doit()
#c
ex = x ** (x**(1/2))
Derivative(ex,x).doit()
#d
ex = (x+3) ** (x**(1/2))
Derivative(ex,x).doit()
#e
ex = sin(x**2) * cos(x)
Derivative(ex,x).doit()
#f
ex = x**(4/5)+sin(x**(1/2))
Derivative(ex,x).doit()
#g
ex = exp(x)*sin(x)
Derivative(ex,x).doit()
#h
ex = log(x**x)
Derivative(ex,x).doit()
```

```python
#12
from sympy import *
#a
x,y,z,a,b,c = symbols('x,y,z,a,b,c')
ex = x**2+6*x+2
Integral(ex).doit()
#b
ex = 2*x**2-6*x+12
Integral(ex).doit()
#c
ex = sin(x+a)
Integral(ex,(x,0,2*pi)).doit()
#d
ex = pi*cos(x+3)
Integral(ex,(x,-1,1)).doit()
#e
ex = y**(1/2)-x**2
Integral(ex,(x,0,3*x),(y,0,1))
#f
ex = y*sin(x)
Integral(ex,(x,0,x**(1/2)),(y,0,pi)).doit()
```
